New column include type of movies

```python
def get_movie_type(Genre):
    if 'Action' in Genre:
        movie_type = 'Action'
    elif 'Drama' in Genre:
        movie_type = 'Drama'
    elif 'Comedy' in Genre:
        movie_type = 'Comedy'
    elif 'Romantic' in Genre:
        movie_type = 'Romantic'
    elif 'Fantasy' in Genre:
        movie_type = 'Fantasy'
    elif 'Animation' in Genre:
        movie_type = 'Animation'
    elif 'Crime' in Genre:
        movie_type = 'Crime'
    elif 'Horror' in Genre:
        movie_type = 'Horror'
    elif 'Adventure' in Genre:
        movie_type = 'Adventure'
    elif 'Sci-Fi' in Genre:
        movie_type = 'Sci-Fi'
    else:
        movie_type = 'Other'

    return movie_type 
```

```python
df['Movie_Type'] = df['Genre'].apply(get_movie_type)
```
